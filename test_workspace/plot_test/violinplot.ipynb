```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
sns.violinplot(data=df, x="deck", y="age", scale_hue=True, hue="alive", split=True, 
               cut=0, gridsize=10000, width=0.9, inner='quartile', palette=['#00AFBB', '#FC4E07'],
               linewidth=2.5, saturation=0.8)
"""
cut:
    Distance, in units of bandwidth size, to extend the density past the extreme datapoints. 
    Set to 0 to limit the violin range within the range of the observed data
inner:
    Representation of the datapoints in the violin interior. If box, draw a miniature boxplot.
    If quartiles, draw the quartiles of the distribution. If point or stick, show each underlying datapoint.
    Using None will draw unadorned violins.
"""
```

```python
sns.violinplot(data=df, x="deck", y="age", hue="alive", split=True, palette=['#00AFBB', '#FC4E07'])
sns.histplot(data=df["deck"], kde=True)
plt.ylim(-50,125)
print(min(df["age"]))
```

```python
sns.boxplot(data=df, x="deck", y="age", hue="alive", palette=['#00AFBB', '#FC4E07'])
sns.histplot(data=df["deck"], kde=True)
plt.ylim(-50,125)
```
