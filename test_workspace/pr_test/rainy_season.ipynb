```python
import numpy as np
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize

def draw(img, vmin=0, vmax=8):
    projection = ccrs.PlateCarree(central_longitude=180)
    img_extent = (-180, 180, -60, 60) # (N5-25, E90-120) <- thailand
    fig = plt.figure()
    ax = plt.subplot(projection=projection)
    ax.coastlines()
    mat = ax.matshow(img,
                     origin='upper',
                     extent=img_extent,
                     vmin=vmin, vmax=vmax,
                     transform=projection)
                     #vmin=vmin, vmax=vmax,
                     #cmap='tab20c'
    fig.colorbar(mat, ax=ax)
    plt.show()
```

```python
path = f"/docker/mnt/d/research/D2/resnet/predictant/main/pr_variance.npy"
variance = np.load(path)
print(variance.shape)

for i in range(12):
    mon = variance[0, i, :, :]
    print(f"month: {i+1}")
    draw(mon, vmin=0, vmax=10)
```
